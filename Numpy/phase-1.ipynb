##Numpy array and basics

```python
import numpy as np
```

creating array from list

```python
arr_1d = np.array([2,34,421,223,])
print("1D array:", arr_1d)

arr_2d = np.array([[12,1,23],[21,2,21]])
print("2D array:", arr_2d)
```

### List vs numpy array

```python
import numpy as np
import time

py_list = [1,2,3]
print("Python list multiplication", py_list * 2)

np_array = np.array([1,2,3])
print("Python array multiplication", np_array * 2)

start = time.time()
py_list = [i*2 for i in range(100000)]
print("\nList operation time: ", time.time() - start)

start = time.time()
np_array = np.arange(100000) * 2
print("\nNumpy operation time: ", time.time() - start)
```

creating array from scratch

```python
zeros = np.zeros((3,4))
print("zero array: \n", zeros)

ones = np.ones((2,3))
print("one array: \n", ones)

full = np.full((2,2), 7)
print("one array: \n", full)

random = np.random.random((2,3))
print("random array: \n", random)

sequence = np.arange(0, 10, 3)
print("sequence array: \n", sequence)
```

Vector, Matrix and Tensor

```python
vector = np.array([1,23,21,2])
print("Vector: ", vector)

matrix = np.array([[1,2,3], [4,5,6]])
print("Matrix: ", matrix)

tensor = np.array([[[1,2], [3,4]],
                   [[5,6], [7,8]]])
print("Tensor: ", tensor)
```

### Array properties

```python
arr = np.array([[1,2,3],[4,5,"6.8"]])
print("Shape ", arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("Dtype ", arr.dtype)
```

Array shipping

```python
arr = np.arange(12)
print("Original array ", arr)

reshaped = arr.reshape((3,4))
print("\n Reshaped array ", reshaped)

flattened = reshaped.flatten()
print("\n flattened array ", flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n reveld array ", raveled)

# transpose
transpose = reshaped.T
print("\n Transpose array ", transpose)
```
