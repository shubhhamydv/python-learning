```python
import numpy as np
import matplotlib.pyplot as plt
print("All good!")
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 220000],
    [2, 120000, 110000, 220000, 220000],
    [3, 200000, 120000, 210000, 250000],
    [4, 180000, 280000, 220000, 210000],
    [5, 160000, 140000, 220000, 150000],
])

print("==== Zomato sales analysis ====")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restau:", sales_data[0:3])
print("\n Sample data for 1st 3 restau:", sales_data[:, 1:])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0))
yeary_total = np.sum(sales_data[:, 1:], axis = 0)
print(yeary_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)

#Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# Average sales per restro
avg_sales = np.mean(sales_data[:, 1:], axis = 1)

```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print(cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumalative sales across all restro")
plt.xlabel("years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([2,4,3,7,5])
print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector", vector1 * vector2)

print("\n Dot product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
