```python
import numpy as np
import pandas as pd
```

### Finding Missing Data

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [24, np.nan, 22, 30],
    "Country": ["USA", "UK", None, "Canada"]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

### Removing Missing Data

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

```python
df
```

```python
df.fillna(0)
```

```python
df
```

```python
values = {'A':0, 'B':200, 'C':300,'D':400}
df.fillna(value=values)
```
